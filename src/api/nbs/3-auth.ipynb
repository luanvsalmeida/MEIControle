## Sign Up

```python
import requests
path = "/api/auth/signup"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"username": "Ian Pereira",
                                                "email": "ian_pereira@mail.com",
                                                "password": "ian123",
                                                "business": "padaria"})
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```
