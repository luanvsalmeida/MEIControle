## Chat

```python
import requests
path = "/api/chat"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"userId": 2, "context": ["vendas do dia"]})
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```

```python
path = "/api/chat/2"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"
response = requests.get(create_endpoint)
print("ok", response.ok)
if response.ok:
    print(response.json())
```
