## Sign Up

```python
import requests
path = "/api/auth/signup"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"username": "Carlos Maria",
                                                "email": "carlos_maria@mail.com",
                                                "password": "maria123",
                                                "business": "confeitaria"})
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```

## Sign In

```python
path = "/api/auth/signin"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"email": "carlos_maria@mail.com",
                                                "password": "maria123"})
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```
